```python
def total(nums: list[int]) ->int:
    total_num:int = 0
    for num in nums:
        total_num += num
    return total_num


print(total([1, 2, 3, 4, 5]))
```

### Error Handling

KeyError, ValueError, ZeroDicisorError

```python
try:
    n=int('abc')
except ValueError:
    print(f"Not a number")
```

```python
try:
    n=int('abc')
except Exception as e:
    print(f"Not a number: {e}")
```

```python
def set_age(age:int) -> None:
    if age < 18:
        raise ValueError("Age must be at least 18")
    print(f"Age is set to {age}")
    
try:
    set_age(15)
except Exception as e:
    print(f"Error: {e}")
```

```python
def safe_int(text:str) -> int:
    try:
        return int(text)
    except ValueError:
        return 0

print(safe_int(42))
print(safe_int("42"))
print(safe_int("abc"))
print(safe_int(""))
```
